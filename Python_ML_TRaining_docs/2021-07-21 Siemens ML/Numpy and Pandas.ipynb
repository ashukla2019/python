```python

p = re.compile(r"\((\d{4})\)")

def extract_year(s):
    m = p.search(s)
    if m:
        return m.group(1)

extract_year("The Gay Desperado (1936)")
```

```python
movies["year"] = movies.title.apply(extract_year)
```
