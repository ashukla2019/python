```python
class Stock:
    def __init__(self, date, symbol, price):
        self.date = date
        self.symbol = symbol
        self.price = price
        
    def __repr__(self):
        return f"{self.symbol} {self.date}"
```

```python
s = Stock("2021-07-21", "GE", 23.0)
s
```
